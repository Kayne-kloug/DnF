```python
def df_search_user(serverId='all', minFame='', maxFame='', jobId='', jobGrowId='', isAllJobGrow='', isBuff='', limit=200):
    if limit >200:
        limit = 200

    url = f'https://api.neople.co.kr/df/servers/{serverId}/characters-fame'
    params = {
        'apikey' : api_key,
        'limit' : limit
    }

    if maxFame != '':
        params['maxFame'] = maxFame
    # 최대 2,000 범위내에서만 검색 가능합니다
    # ※ minFame : 50000, maxFame : 60000 요청 시 minFame : 58000, maxFame : 60000으로 변경되어 검색 결과 제공
    if minFame != '':
        params['minFame'] = minFame

    if jobId != '':
        params['jobId'] = jobId
    if minFame != '':
        params['jobGrowId'] = jobGrowId
    if minFame != '':
        params['isAllJobGrow'] = isAllJobGrow
    if minFame != '':
        params['isBuff'] = isBuff

    return requests.get(url, params=params)
```

```python
output =pd.DataFrame()
def solve_dict(index, solve, upper_key=''):
    global output
    for key, value in solve.items():
        if key=='slotId':
            upper_key=value+'_'+upper_key
        
        if type(value) ==dict:
            upper_key = upper_key+'_'+key
            solve_dict(index, value, upper_key)
        elif type(value) == list:
            for n, slot in enumerate(value):
                if n==0:
                    k=1
                upper_key = upper_key+'_'+key
                solve_dict(index, slot, upper_key)
                upper_key = upper_key.replace('_'+key,'')
        else:
            if upper_key =='':
                output.loc[index,key]=value
            else:
                output.loc[index,upper_key+'_'+key]=value
```
